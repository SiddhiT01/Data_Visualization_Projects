```python
import altair as alt
import pandas as pd

dataset = pd.read_csv('/content/sample_data/supermarket_sales - Sheet1.csv') #load the dataset
dataset.head() # To print the dataset in table form
```

```python
import pandas as pd
import altair as alt


#calculating the total sales amount
total_sales_by_city_product = dataset.groupby(['City', 'Product line'])['Total'].sum().reset_index()

# Create the grouped bar chart
grouped_bar_chart = alt.Chart(total_sales_by_city_product).mark_bar().encode(
    x='City',
    y='Total',
    column='Product line',  # Separating bars by product line
    color='Product line' , # Adding colors
    tooltip=['Product line', 'Total', 'City'] #Adding tooltip when hover
).properties(
    width=100,
    height=300,
    title='Total Sales Amount by Product Line in Each City'
)
# Show the grouped bar chart
grouped_bar_chart
```
